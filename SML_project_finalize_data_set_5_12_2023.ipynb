```python

# Calculate the IQR for the 'visibility' column
Q1 = df['windspeed'].quantile(0.25)
Q3 = df['windspeed'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds to identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers from the 'visibility' column
df_no_outliers = df[(df['windspeed'] >= lower_bound) & (df['windspeed'] <= upper_bound)]

# Display the DataFrame without outliers
print(df_no_outliers)
```

```python
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
df_no_outliers['increase_stock'] = encode.fit_transform(df_no_outliers['increase_stock'])
df_no_outliers
```

```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

# Assuming df_no_outliers is your DataFrame

# Separating features (X) and target variable (y)
X = df_no_outliers.drop('increase_stock', axis=1)
y = df_no_outliers['increase_stock']

# Instantiate SMOTE with the desired sampling strategy
smote = SMOTE(sampling_strategy='minority')

# Apply SMOTE to the dataset
X_sm, y_sm = smote.fit_resample(X, y)


# Create a new DataFrame with the balanced data
df_balanced = pd.concat([pd.DataFrame(X_sm, columns=X.columns), pd.Series(y_sm, name='increase_stock')], axis=1)

# Display the count of each class in the target variable after applying SMOTE
print("Class distribution after SMOTE:")
print(df_balanced['increase_stock'].value_counts())

df_balanced
```

```python
from sklearn import preprocessing

X = df_balanced.drop('increase_stock', axis=1)
y =  df_balanced.increase_stock

min_max_scaler = preprocessing.MinMaxScaler()
X = min_max_scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.25, random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
fitted_model = model.fit(X_train, Y_train)
model_prediction = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score
print("Accuracy Score : ", accuracy_score(Y_test, model_prediction))
print("Misclassification error: ", 1 - accuracy_score(Y_test, model_prediction))
```

```python
"""KNN prediction is evaluated on the remaining 25% of the data."""
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix


knn = skl_nb.KNeighborsClassifier(n_neighbors=4, metric='minkowski',
      weights='distance')
knn.fit(X_train, Y_train)
predicted = knn.predict(X_test)

acc_score = accuracy_score(Y_test, predicted)
precision = precision_score(Y_test, predicted)

cm = confusion_matrix(Y_test, predicted)
sensitivity = cm[0,0]/(cm[0,0]+cm[0,1])
specificity = cm[1,1]/(cm[1,0]+cm[1,1])

print("Accuracy score ",acc_score)
print("precision",precision)
print("sensitivity",sensitivity)
print("specificity",specificity)
```
